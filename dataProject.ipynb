```python
city_attributes = pd.read_csv("city_attributes.csv")
humidity = pd.read_csv("humidity.csv")
pressure = pd.read_csv("pressure.csv")
temperature = pd.read_csv("temperature.csv")
weather_description = pd.read_csv("weather_description.csv")
wind_direction = pd.read_csv("wind_direction.csv")
wind_speed = pd.read_csv("wind_speed.csv")
```

```python
list = [humidity,pressure,temperature,weather_description,wind_direction,wind_speed]
for i in list:
    i.datetime= i.datetime.apply(pd.to_datetime)
```

```python
humidity["month"]=humidity.datetime.dt.month
humidity["year"]=humidity.datetime.dt.year

pressure["month"]=pressure.datetime.dt.month
pressure["year"]=pressure.datetime.dt.year

temperature["month"]=temperature.datetime.dt.month
temperature["year"]=temperature.datetime.dt.year

weather_description["month"]=weather_description.datetime.dt.month
weather_description["year"]=weather_description.datetime.dt.year

wind_direction["month"]=weather_description.datetime.dt.month
wind_direction["year"]=weather_description.datetime.dt.year

wind_speed["month"]=wind_speed.datetime.dt.month
wind_speed["year"]=wind_speed.datetime.dt.year
```

```python

broken_clouds = weather_description[weather_description[:]=="broken clouds"].count()
sky_is_clear = weather_description[weather_description[:]=="sky is clear"].count()
scattered_clouds = weather_description[weather_description[:]=="scattered clouds"].count()
few_clouds = weather_description[weather_description[:]=="few clouds"].count()
overcast_clouds = weather_description[weather_description[:]=="overcast_clouds"].count()
light_rain = weather_description[weather_description[:]=="light_rain"].count()
haze = weather_description[weather_description[:]=="haze"].count()
mist = weather_description[weather_description[:]=="mist"].count()

```
